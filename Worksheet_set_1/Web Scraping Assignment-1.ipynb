### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
imdb100movies_page = requests.get("https://www.imdb.com/list/ls091520106/")
imdb100movies_page_content = BeautifulSoup(imdb100movies_page.content)
year_of_release = []
for i in imdb100movies_page_content.find_all('span',class_="lister-item-year text-muted unbold"):
    year_of_release.append(i.text.replace('(','').replace(')','').replace('I','').replace(" ",''))
#year_of_release

name_of_movies = []
for j in imdb100movies_page_content.find_all('h3',class_="lister-item-header"):
    name_of_movies.append(j.a.text)
#name_of_movies

rating = []
for k in imdb100movies_page_content.find_all('div',class_="ipl-rating-star small"):
    rating.append(k.text.replace('\n',''))
#rating

df = pd.DataFrame({'Name':name_of_movies,'Rating':rating,'Year':year_of_release})
df
```

### 5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
##### a. Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
odi_team_page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
odi_team_page_content = BeautifulSoup(odi_team_page.content)

odi_team = []
for i in odi_team_page_content.find_all('span',class_="u-hide-phablet"):
    odi_team.append(i.text)
odi_team = odi_team[:10]

rat = odi_team_page_content.find('td',class_="rankings-block__banner--rating u-text-right").text.replace("\n","").strip()
odi_rating = [rat]
for i in odi_team_page_content.find_all('td',class_="table-body__cell u-text-right rating"):
    odi_rating.append(i.text.replace('\n','').replace(' ',''))
odi_rating = odi_rating[:10]

first_rec  = odi_team_page_content.find('td',class_='rankings-block__banner--matches').text

matches = []
for j in odi_team_page_content.find_all('td',class_= "table-body__cell u-center-text"):
        matches.append(j.text)
#matches

point = matches[1::2]
point = point[:10]

mat = matches[::2]
mat.insert(0,first_rec)
mat = mat[:10]

Top_10_mens_ODI = pd.DataFrame({})
Top_10_mens_ODI['Team_Name'] = odi_team[:10]
Top_10_mens_ODI['Matches'] = mat
Top_10_mens_ODI['Points'] = point
Top_10_mens_ODI['Ratings'] = odi_rating[:10]
Top_10_mens_ODI
```

### 7. Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
coreyms_page = requests.get("https://coreyms.com/")
coreyms_page_content = BeautifulSoup(coreyms_page.content)

headings = []
for i in coreyms_page_content.find_all('a',class_="entry-title-link"):
    headings.append(i.text)

times = []
for i in coreyms_page_content.find_all('time',class_="entry-time"):
    times.append(i.text)

contents = []
for i in coreyms_page_content.find_all('div',class_="entry-content"):
    contents.append(i.text.replace('\n',''))

youtube_link = []
for i in coreyms_page_content.find_all('iframe',class_="youtube-player"):
    youtube_link.append(i['src'])
youtube_link.insert(4,"Video not available")

coreyms_df = pd.DataFrame({"Heading":headings,"Time":times,"Content":contents,"YoutubeLink":youtube_link})
coreyms_df
```

### 9. Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
dineout_page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
dineout_content = BeautifulSoup(dineout_page.content)

rest_name = []
for i in dineout_content.find_all("a",class_="restnt-name ellipsis"):
    rest_name.append(i.text)

csn = []
for i in dineout_content.find_all('span',class_="double-line-ellipsis"):
    csn.append(i.text.split('|')[1].strip())

loc = []
for i in dineout_content.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)

img_url = []
for i in dineout_content.find_all('img',class_="no-img"):
    img_url.append(i["data-src"])

dine_rating = []
for i in dineout_content.find_all("div",class_="restnt-rating rating-4"):
    dine_rating.append(i.text)

dineout_df = pd.DataFrame({"Restaurant name":rest_name,"Cuisine":csn,"Location":loc,"Rating":dine_rating,"Image URL":img_url})
dineout_df
```
