```python
ratings = []

rat = driver.find_elements_by_xpath("//div[@class='score orange big']")

for i in rat:
    ratings.append(i.text)
    
ratings
```

```python
description = []

dis = driver.find_elements_by_xpath("//div[@class='keyword']")

for i in dis:
    try:
        description.append(i.text)
    except:
        description.append("-")
description
```
